```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
url = 'https://www.rightmove.co.uk/property-for-sale/find.html?searchType=SALE&locationIdentifier=OUTCODE%5E251&insId=1&radius=0.0&minPrice=&maxPrice=&minBedrooms=&maxBedrooms=&displayPropertyType=&maxDaysSinceAdded=&_includeSSTC=on&sortByPriceDescending=&primaryDisplayPropertyType=&secondaryDisplayPropertyType=&oldDisplayPropertyType=&oldPrimaryDisplayPropertyType=&newHome=&auction=false'
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
body = soup.find('div', class_= 'main')
search = body.find('div', class_ = 'l-searchResults')
print(search)
```

```python
propert = search.find_all('div', class_="l-searchResult is-list")
propert
```

```python
total = []
for i in propert:
    proper = i.find('div', class_ = 'propertyCard-wrapper')
    total.append(proper)
total
    
```

```python
total2 = []
for i in total:
    prope = i.find('div', class_ = 'propertyCard-section').find('div', class_ = 'propertyCard-details')
    total2.append(prope)
total2
```

```python
address1 = []
for i in total2:
    address = i.find('address', class_ = 'propertyCard-address').text
    address1.append(address)
address1
```

```python
names1 = []
for i in total2:
    name = i.find('h2', class_ = 'propertyCard-title').text
    names1.append(name)
names1
```

```python
total3 = []
for i in total:
    prope = i.find('div', class_ = 'propertyCard-section')
    total3.append(prope)
total3
```

```python
descriptions = []
for i in total3:
    discrip = i.find_all('span')[1].text
    descriptions.append(discrip)
descriptions
```
